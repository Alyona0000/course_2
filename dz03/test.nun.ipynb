```python
import numpy as np
# несколоко колод карт

num_players = 4   # кількість гравців
num_cards = 3     #кількість розданих карт 
num_decks = 5

num_decks = np.array([np.arange(7,15)+100, np.arange(7,15)+200, np.arange(7,15)+300, np.arange(7,15)+400])#двумерний масив с рартами по масти
num_decks = num_decks.flatten()#плоский масив с колодой картю
s = np.random.permutation(num_decks)
print(s)


for i in range(num_players):
    player = s[:num_cards]   # беремо перші k карт
    s = s[num_cards:]        # зменшуємо колоду
    print(f"Гравець {i+1} отримав:", player)
    print("Залишок у колоді:", s)
```
